Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

we are having 807 fraud transactions so we take 807 nt=rml transactions and combine tehm with fraud transactions.


Here fraud transactions are very less comoared to nrml ones

```python
legit_sample = legit.sample(n=807)
```

concatinating two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
numeric_cols = new_dataset.select_dtypes(include='number').columns
new_dataset[numeric_cols].groupby(new_dataset['Class']).mean()
```

splitting the data into features and targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

tarining and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

model training

LOgistic Regression

```python
 model = LogisticRegression()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

model evaluation

accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
